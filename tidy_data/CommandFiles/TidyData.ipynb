## Original Data ##

```python
import pandas as pd

df = pd.read_csv('billboard.csv', encoding_errors='ignore')
df
```

# Show info about data #

```python
df.info()
```

# Create sub-dataframes #

```python
subdf_weeks = df.drop(['year', 'time', 'genre', 'date.entered', 'date.peaked', 'artist.inverted', 'track'], axis=1)
subdf_tracks = df[['artist.inverted', 'track']]
subdf_weeks
```

```python
subdf_tracks
```

# Tidy data #

```python
df2 = pd.DataFrame(columns=['artist.inverted', 'track', 'week', 'rank'])
df3 = pd.DataFrame(columns=['artist.inverted', 'track', 'weeks on charts'])
counter = 0
for i in range(0,316):
    counter = 0
    for j in range(0,75):
        if pd.isna(subdf_weeks.iat[i,j]) is False:
            new_row2 = {'artist.inverted':subdf_tracks.iat[i,0], 'track':subdf_tracks.iat[i,1], 'week':j+1, 'rank':subdf_weeks.iat[i,j]}
            df2 = df2.append(new_row2, ignore_index=True)
            counter +=1
    new_row3 = {'artist.inverted':subdf_tracks.iat[i,0], 'track':subdf_tracks.iat[i,1], 'weeks on charts':counter}
    df3 = df3.append(new_row3, ignore_index=True)
```

# Save dataframes #

```python
df2.to_csv('billboards_weeks.csv')
df3.to_csv('billboards_counter.csv')
```
